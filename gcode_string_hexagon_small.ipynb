```python
def write_header_lines(file, layer_height, line_width, layer_count, mesh, speed=1200):
    with open(file, 'w') as f:
        header_lines = [
            f";layer_height = {layer_height}",
            f"\n;line_width = {line_width}",
            f"\n;layer_count = {layer_count}",
            f"\n;mesh = {mesh}"
        ]
        
        initialize_lines = [
            "\nG21 ;start of the code",
            "\nG1 Z15 F300",
            "\nG28 X0 Y0 ;Home",
            "\nG92 X0 Y0 ;Consider this as current",
            "\nG0 X50 Y50 F3000 ;Go-to Offset",  
            "\nG92 X0 Y0 ;Reset",
            "\n",
            "\nG0 F3600 X-8.26 Y-14.31 Z0.3",
            "\nM7",
            "\n\n"
        ]
        
        f.writelines(header_lines)
        f.writelines(initialize_lines)
        
        
def writ_finish_lines(file):
    with open(file, 'a') as f:
        finish_lines = [
            "\n\n;Finish",
            "\nM9",
            "\nG28 X-100 Y-100;Home",
        ]
        f.writelines(finish_lines)
```

```python
def rotate_60_degree(xy):
    x, y = xy
    x_new = x * np.cos(np.pi/3) - y * np.sin(np.pi/3)
    y_new = x * np.sin(np.pi/3) + y * np.cos(np.pi/3)
    return np.array([x_new, y_new])

def rotate_xy(xy, times):
    for _ in range(times):
        xy = rotate_60_degree(xy)
    return xy

def write_one_line_with_rotate(lines, xy, rotate_times):
    xy = rotate_xy(xy, rotate_times)
    x, y = xy
    lines.append(f"\nG1 X{x:.3f} Y{y:.3f}")
    # return lines
```

```python
def draw_layer(file, x_barrier):
    lines = []
    
    for i in range (6):
        xy = np.array([-7.34, -12.71])
        if (x_barrier > -6.74) and (x_barrier < 6.74):
            xy += [7.34+x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34+x_barrier-0.6),0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [7.34+x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34+x_barrier-0.6),0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [14.68, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34-x_barrier-0.6), 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [7.34-x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34-x_barrier-0.6), 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [7.34-x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            
        elif x_barrier <= -6.74:
            xy += [0, -1.6]
            write_one_line_with_rotate(lines, xy, i)
            xy += [14.68, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34-x_barrier-0.6), 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [7.34-x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34-x_barrier-0.6), 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [7.34-x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            
        elif x_barrier >= 6.74:
            xy += [7.34+x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34+x_barrier-0.6),0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [7.34+x_barrier-0.6, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [-(7.34+x_barrier-0.6),0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, -0.4]
            write_one_line_with_rotate(lines, xy, i)
            xy += [14.68, 0]
            write_one_line_with_rotate(lines, xy, i)
            xy += [0, 1.6]
            write_one_line_with_rotate(lines, xy, i)
            
    with open(file, 'a') as f:
        f.writelines(lines)
```

```python
def write_bottom_layer(file):
    z = 0.3
    
    lines = []
    xy = np.array([-8.26, -14.31])
    xy += [16.52, 0]
    for i in range(6):
        write_one_line_with_rotate(lines, xy, i)
    xy -= [16.52, 0]
    
    l = 16.06
    while(l > 0):    
        xy += [0.23, 0.4]
        write_one_line_with_rotate(lines, xy, 0)
        xy += [l, 0]
        for i in range(6):
            write_one_line_with_rotate(lines, xy, i)
        xy -= [l, 0]
        l -= 0.46
    
    lines.append(f"\nG1 X0 Y0")
    z += 0.4
    lines.append(f"\nG1 Z{z}")
    
    length = len(lines)
    for i in range (length - 1):
        lines.append(lines[length - i - 2])
        
    lines.append(f"\nG1 X-8.26 Y-14.31")
    z += 0.4
    lines.append(f"\nG1 Z{z}")
    lines.append(f"\nG1 X-7.34 Y-12.71")
    
    # lines.append(f"\nG1 X{xy[0]:.3f}, Y{xy[1]:.3f}")
    # while(l <= 16.52):
    #     xy -= [0.23, 0.4]
    #     write_one_line_with_rotate(lines, xy, 0)
    #     xy += [l, 0]
    #     for i in range(6):
    #         write_one_line_with_rotate(lines, xy, i)
    #     xy -= [l, 0]
    #     l += 0.46
    
    with open(file, 'a') as f:
        f.writelines(lines)
```
